```python
import numpy as np
from collections import Counter
```

```python
class Node:
  def __init__(self,feature=None,threhold=None,left=None,right=None,value=None):
    self.feature=feature
    self.threhold=threhold
    self.left=left
    self.right=right
    self.value=value
```

```python
class DecisionTree:
  def __init__(self,min_sample_split=2,max_depth=100,n_features=None):
    self.min_sample_split=min_sample_split
    self.max_depth=max_depth
    self.n_features=n_features
    self.root=None
  def fit(self,X,y):
    self.n_features=X.shape[1] if self.n_features is None else min(self.n_features,X.shape[1])
    self.root=self.build_tree(X,y)
  def build_tree(self,X,y,depth=0):
    n_sample,n_features=X.shape
    n_labels=len(np.unique(y))
    if depth>=self.max_depth or n_labels==1 or n_sample<self.min_sample_split:
      leaf=self.majority_vote(y)
      return Node(value=leaf)
    feature,threhold=self.best_split(X,y)
    left_ind,right_ind=self.split(X[:,feature],threhold)
    left=self.build_tree(X[left_ind,:],y[left_ind],depth+1)
    right=self.build_tree(X[right_ind,:],y[right_ind],depth+1)
    return Node(feature,threhold,left,right,value=None)
  def best_split(self,X,y):
    best_gain=-1
    n_sample,n_features=X.shape
    best_ind=None
    best_threhold=None
    for feature in n_features:
      X_col=X[:,feature]
      threholds=np.unique(X_col)
      for threhold in threholds:
        gain=self.information_gain(y,X_col,threhold)
        if gain>best_gain:
          best_gain=gain
          best_ind=feature
          best_threhold=threhold
    return best_ind,best_threhold
  def best_split(self,X,y):
    best_gain=-1
    n_sample,n_features=X.shape
    best_ind=None
    best_threhold=None
    for feature in range(n_features):
      X_col=X[:,feature]
      threholds=np.unique(X_col)
      for threhold in threholds:
        gain=self.information_gain(y,X_col,threhold)
        if gain>best_gain:
          best_gain=gain
          best_ind=feature
          best_threhold=threhold
    return best_ind,best_threhold

  def information_gain(self,y,X_col,threhold):
    parent_entropy=self.entropy(y)
    left_ind,right_ind=self.split(X_col,threhold)
    if len(left_ind)==0 or len(right_ind)==0:
      return 0
    n,n_left,n_right=len(y),len(left_ind),len(right_ind)
    e_l=self.entropy(y[left_ind])
    e_r=self.entropy(y[right_ind])
    child_entropy=(n_left/n)*e_l+(n_right/n)*e_r
    information_gain=parent_entropy-child_entropy
    return information_gain
  def split(self,X_col,threhold):
    left_ind=np.argwhere(X_col<=threhold).flatten()
    right_ind=np.argwhere(X_col>threhold).flatten()
    return left_ind,right_ind
  def entropy(self,y):
    hist=np.bincount(y)
    ps=hist/len(y)
    return -np.sum([np.log2(p) for p in ps if p>0])
  def majority_vote(self,y):
    count=Counter(y)
    return count.most_common(1)[0][0]
  def predict(self,X):
    return np.array([self.traverse_tree(x,self.root) for x in X])
  def traverse_tree(self,x,node):
    if node.value is not None:
      return node.value
    if x[node.feature]<=node.threhold:
      return self.traverse_tree(x,node.left)
    return self.traverse_tree(x,node.right)




```

```python


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data=load_breast_cancer()
```

```python
x,y=data.data,data.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=DecisionTree(min_sample_split=2, max_depth=10)
model.fit(X_train,y_train)
```

```python
test_pred=model.predict(X_test)
```

```python
accurracy=np.sum(test_pred==y_test)/len(y_test)*100
```
