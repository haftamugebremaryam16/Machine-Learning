```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import matplotlib.pyplot as plt
```


###I Will implement logistic regression from scratch using numpy only






$$
\text{loss} = -\left[y \log(\hat{y}) + (1 - y)\log(1 - \hat{y})\right]
$$

$$
\text{cost} = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)}) \right]
$$

```python
def z_score(x):
  mu=np.mean(x,axis=0)
  std=np.std(x,axis=0)
  return mu,std,(x-mu)/std
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def compute_cost(x,y,w,b):
  m=x.shape[0]
  z=np.dot(x,w)+b
  y_hat=sigmoid(z)
  cost=-(1/m)*np.sum(y*np.log(y_hat)+(1-y)*np.log(1-y_hat))
  return cost

```

```python
def compute_gradient(x,y,w,b):
  m=x.shape[0]
  z=np.dot(x,w)+b
  y_hat=sigmoid(z)
  dw=(1/m)*np.dot(x.T,(y_hat-y))
  db=(1/m)*np.sum(y_hat-y)
  return dw,db
```

```python
def gradinet_descent(x,y,w,b,alpha,iterations):
  m=x.shape[0]
  cost=[]

  for i in range(iterations):
    dw,db=compute_gradient(x,y,w,b)
    w=w-alpha*dw
    b=b-alpha*db
    cost.append(compute_cost(x,y,w,b))
  return w,b,cost
```

```python
data=load_breast_cancer()
```

```python
X_train=data['data']
Y_train=data['target']
```

```python
w=np.zeros(X_train.shape[1])
b=0
alpha=0.01
iterations=1000
mu,std,X_norm=z_score(X_train)
```

```python
w,b,cost=gradinet_descent(X_norm,Y_train,w,b,alpha,iterations)
```

```python
plt.plot(cost)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Iterations')
plt.show()
```

```python
z=np.dot(X_norm,w)+b
y_pred=sigmoid(z)
```

```python
y_mean=np.mean(y_pred)
```

```python

y_pred_labels = (y_pred >= 0.5).astype(int)
accuracy = np.mean(y_pred_labels.flatten() == Y_train.flatten())
print(f"Accuracy: {accuracy:.4f}")
```

```python
from sklearn.metrics import precision_score,recall_score

precision = precision_score(Y_train, y_pred_labels)
recall=recall_score(Y_train, y_pred_labels)
print("Precision:", precision)
print("recall:",recall)
```
