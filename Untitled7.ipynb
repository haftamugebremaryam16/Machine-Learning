```python
import numpy as np
```

```python
class K_mean:
  def __init__(self, k=2, tol=0.001, max_iter=300):
    self.k = k
    self.tol = tol
    self.max_iter = max_iter
    self.centroids = None
  def fit(self, X):
    centroids = X[np.random.choice(X.shape[0],self.k,replace=False)]
    for i in range(self.max_iter):
      ind = self.closest_path(X, centroids)
      self.centroids = self.update_centroids(X, ind, centroids)
    return self
  def closest_path(self,X,centroids):
    m,n=X.shape
    ind=np.zeros(m)
    for i in range(m):
      dis=[np.linalg.norm(X[i]-centroids[j]) for j in range(self.k)]
      ind[i]=np.argmin(dis)

    return ind
  def update_centroids(self,X,ind,centroids):
      m,n=X.shape

      for j in range(self.k):
        point=X[ind==j]
        if len(point)>0:
          centroids[j]=np.mean(point,axis=0)
      return centroids

  def compute_cost(self,X,centroids,ind):
    m=X.shape[0]
    cost=0
    for i in range(m):
      cost+=np.sum((X[i]-centroids[ind[i]])**2)
    return cost/m
  def predict(self,X):
    m=X.shape[0]
    labels=np.zeros(m,dtype=int)
    for i in range(m):
      distance=[np.linalg.norm(X[i]-self.centroids[j]) for j in range(self.k)]
      labels[i]=np.argmin(distance)
    return labels


```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs()
```

```python
clusters=len(np.unique(y))
```

```python
model=K_mean(k=clusters,max_iter=100)
```

```python
model.fit(X)
```

```python
model.predict(X)
```
